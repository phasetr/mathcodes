▲リスト3.11：2次の多項式をグラフで描画する

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

def my_func(x):  
    return 3*x**2 + 2*x + 1

x = np.linspace(-2, 2)
y = my_func(x)  # y = f(x)

plt.plot(x, y)
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```
