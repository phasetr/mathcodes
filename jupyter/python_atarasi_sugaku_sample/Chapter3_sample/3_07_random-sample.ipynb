▲リスト3.23：1から6までの、整数の乱数を生成する

```python
import numpy as np

r_int = np.random.randint(6) + 1  # 0から5までの乱数に、1を加える
print(r_int)  # 1から6までがランダムに表示される
```

▲リスト3.24：0から1までの、小数の乱数を生成する

```python
import numpy as np

r_dec = np.random.rand()   # 0から1の間の小数を、ランダムに返す
print(r_dec)
```

▲リスト3.25：乱数の均一な分布

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

n = 1000  # サンプル数
x = np.random.rand(n)  # 0-1の均一な乱数
y = np.random.rand(n)  # 0-1の均一な乱数

plt.scatter(x, y)  # 散布図のプロット
plt.grid()
plt.show()
```

▲リスト3.26：正規分布に従う乱数の分布

```python
import numpy as np
import matplotlib.pyplot as plt

n = 1000  # サンプル数
x = np.random.randn(n)  # 正規分布に従う乱数
y = np.random.randn(n)  # 正規分布に従う乱数

plt.scatter(x, y)  # 散布図のプロット
plt.grid()
plt.show()
```
