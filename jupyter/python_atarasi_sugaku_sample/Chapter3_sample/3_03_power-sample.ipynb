▲リスト3.7：べき乗のグラフを描画する

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

def my_func(x):
    a = 3
    return x**a  # xのa乗

x = np.linspace(0, 2)
y = my_func(x)  # y = f(x)

plt.plot(x, y)
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```

▲リスト3.8：平方根のグラフを描画する

```python
import numpy as np
import matplotlib.pyplot as plt

def my_func(x):
    return np.sqrt(x)  # xの正の平方根。x**(1/2) でも同じ

x = np.linspace(0, 9)
y = my_func(x)  # y = f(x)

plt.plot(x, y)
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```
