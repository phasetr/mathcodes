▲リスト7.5：入力データと正解データを用意する

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

# -- 入力と正解の用意 --
X = np.linspace(-np.pi/2, np.pi/2)  # 入力: -π/2からπ/2の範囲
T = (np.sin(X) + 1)/2  # 正解: 0から1の範囲
n_data = len(T)  # データ数

# --- グラフで描画してみる ---
plt.plot(X, T)

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
```
