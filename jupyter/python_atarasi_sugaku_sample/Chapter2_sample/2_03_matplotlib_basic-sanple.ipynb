▲リスト2.51：pyplotでシンプルなグラフを描画する

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5)  # -5から5まで
y = 2 * x  # xに2を掛けてy座標とする

plt.plot(x, y)
plt.show()
```

▲リスト2.52：グラフを装飾する

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5)
y_1 = 2 * x
y_2 = 3 * x

# 軸のラベル
plt.xlabel("x value", size=14)  # 軸ラベルの文字の大きさを14に指定
plt.ylabel("y value", size=14)

# グラフのタイトル
plt.title("My Graph")

# グリッドの表示
plt.grid()

# プロットの際に凡例と線のスタイルを指定
plt.plot(x, y_1, label="y1")
plt.plot(x, y_2, label="y2", linestyle="dashed")
plt.legend() # 凡例を表示

plt.show()
```

▲リスト2.53：scatterで散布図を表示する

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1.2, 2.4, 0.0, 1.4, 1.5, 0.3, 0.7])  # x座標
y = np.array([2.4, 1.4, 1.0, 0.1, 1.7, 2.0, 0.6])  # y座標

plt.scatter(x, y)  # 散布図のプロット
plt.grid()
plt.show()
```
