# 数列とは何か? 微分方程式のシミュレーションの観点から

## 数列も関数
ベクトル、関数と来て今回は数列です。
簡単に復習すると、
関数は数と数の割り当てルールのことで、数列も関数なのでした。

数列も関数となると、
数と数をどう割り当てるのかというルール指定が必要です。
そのルールという視点に注目した話をしていきます。

## 等比数列
世にはいろいろな数列があります。
高校でもいろいろ出てきます。
今回の話で 1 番大事な数列は**等比数列**です。
長々とやっても仕方ないので軽く復習しておきます。

数列 $(x_n)$ を考えましょう。

（**注**：高校だと $\{x_n\}$ と中括弧で書くと思いますが、
大学の数学的な都合で小括弧で書くことにします。
単なる記法の問題なので必要以上に気にする必要はありません。）

等比数列がどんな数列だったか、
つまりどんなルールで与えられているのかというとこんなルールです。

\begin{align}
 x_{n+1}
 =
 \alpha x_{n}.
\end{align}

$x_{n}$ が 0 のときでも成り立つ形で書きました。
$x_{n}$ が 0 ではないならもちろん $x_{n+1} / x_{n} = \alpha$ と書けます。
どんな $n$ に対しても比が一定な数列だから等比数列というのでした。

## 等比数列の一般項
このルールを使って**一般項**を出します。
次のように延々と続きます。

- $x_{n+1} = \alpha x_{n}$,
- $x_{n} = \alpha x_{n-1}$,
- $x_{n-1} = \alpha x_{n-2}$.

これが終わるのは $x_1 = \alpha x_{0}$ です。
順々に代入していくと $x_{n} = x_{0} \alpha^{n}$ になります。

この計算、ハマる人はハマります。
実際私も高校の頃、時々わけがわからなくなることがありました。
試験中に慌ててパニックになったことを思い出します。

それはそれとして、
ここで細々とした計算を詳しく解説していると本筋を見失うので省略します。
数列をどんなところでどう使うのか、
それをちゃんと見るのがこの講座の目的だからでもあります。

## 微分方程式と等比数列
これがどこで出てきたかというと、微分方程式を近似した方程式で出てきたのでした。
放射性物質の崩壊で出てきたのはこんな式です。

\begin{align}
 \frac{x_{n+1} - x_{n}}{h}
 =
 - x_{n}。
\end{align}

これを整理すると $x_{n+1} = (1 - h) x_{n}$ で、
$1 - h = \alpha$ とすればまさに等比数列です。
関数を決めるルールとしての微分方程式というイメージはここから作ってみてください。

先程書いたように等比数列は**指数関数**で書けます。
そして元の微分方程式の解も実は指数関数で書けます。

## もう一度: 数列も関数
少し先走ったことも書きました。
何はともあれ、数列は関数であり、関数は数と数の割り当てルールのことで、
その割り当てルールとして等比数列という割り当て方があるのです。
微分方程式もその割り当てルールと見なせるのだというのがいま説明したことです。
微分方程式と揃えた言い方として上の式は**差分方程式**と呼ぶこともあります。

## 漸化式
ちなみに上で紹介した割り当てルールはもっと一般的にできます。
要は次の項はその前の項で決まると言っているだけなので、
$a_{n+1} = a_{n}^2$ みたいにしてもよく、
$a_{n+1} = a_{n} + a_{n-1}$ のように 2 つ前と関係があるとしても構いません。
これを一般化したのがいわゆる**漸化式**です。

## 2 つの数列に対する漸化式
もちろんこれ以外にもいろいろな漸化式があります。
あとで使うのでもう 1 つ漸化式を紹介します。
そのために 2 つの数列 $(x_{n})$、 $(v_{n})$ を用意します。
そして次のような漸化式を考えます。

\begin{align}
 x_{n+1} - x_{n}
 =
 v_{n}, \quad
 v_{n+1} - v_{n}
 =
 - \alpha^2 x_{n}.
\end{align}

2 つの数列が絡まりあう漸化式です。
高校の言葉を使うなら、
互いに**階差数列**を取り合う数列です。

## 単振動の漸化式
これがどこから出てくるかというと次の微分方程式からです。

\begin{align}
 \frac{d^2 x (t)}{dt^2}
 =
 - \omega^{2} x (t).
\end{align}

この微分方程式も物理でよく出てくる方程式で、
いわゆる**単振動の運動方程式**です。
さっきの漸化式を出すには $v (t) = x'(t)$ としてもとの方程式も $v' = - \omega^2 x$ とします。
これを差分近似すると次の式が出てきます。

\begin{align}
 x_{n+1}
 =
 x_{n} + h v_{n}, \quad
 v_{n+1}
 =
 v_{n} - h \omega^{2} x_{n}.
\end{align}

$h$ が入っていて係数が微妙に違いますが、
あまり気にしないでください。

## 単振動の漸化式を解いてみる
放射性元素の崩壊の微分方程式とその近似解はすでに紹介済みです。
ここでは改めて単振動の漸化式を解いてみましょう。
初回で既に紹介したプログラムではありますが、「大事なことは何度でも」の精神で再掲します。

## オイラー法による定式化
一番最初に書いたように、上の「素直な」定式化は近似の精度に問題があります。

まずは細かいことを気にせず何となくは解けていることに注目して読み進めてください。

```python
import numpy as np
import matplotlib.pyplot as plt

def harmonic_euler(nt, init = (5, 0)):
    dt = t_range / (nt - 1)
    x = np.zeros(nt)
    v = np.zeros(nt)
    # 初期条件設定
    x[0] = init[0]
    v[0] = init[1]

    for i in range(1, nt):
        # ここで方程式を解いている
        x[i] = x[i-1] + dt * v[i-1]
        v[i] = v[i-1] - dt * (omega ** 2) * x[i-1]

    return (x, v)

# パラメータ設定
omega = 2 * np.pi
nt = 101
t_range = 2
init = (5, 0)

# 近似解
harm = harmonic_euler(nt, init)
t = np.linspace(0, t_range, nt)

# 厳密解
x_exact = init[0] * np.cos(omega * t)
v_exact = - omega * init[0] * np.sin(omega * t)

# グラフ描画
plt.subplot(3, 1, 1) # 縦に 3 つグラフを並べていて、そのうちの一番上
plt.title('x-t graph')
plt.plot(t, harm[0]) # 近似解
plt.plot(t, x_exact) # 厳密解
plt.legend(['x approximation', 'x exact'])

plt.subplot(3, 1, 2)
plt.title('v-t graph')
plt.plot(t, harm[1]) # 近似解
plt.plot(t, v_exact) # 厳密解
plt.legend(['v approximation', 'v exact'])

plt.subplot(3, 1, 3)
plt.title('phase space')
plt.plot(harm[0], harm[1])

# 描画
plt.tight_layout()
```

## グラフの見方
見ての通り時間が進むごとに誤差が大きくなります。
実は `nt` を大きくすると少しはましになります。
実際に上のコードで `nt` を大きくして再計算してみてください。

物理的に一番問題なのは一番下の phase space （相空間）の図です。
この系はエネルギーが保存する系なので、相空間内の軌道が閉じてほしいのにそうなっていません。
シンプレクティックにやれば解消できる部分もありますが、
とにかくここではよろしくない挙動です。

もっと検討すべき項目はありますが、
ここではまずこの方程式を**素直に**オイラー法で解くと問題が出ることに注目しましょう。
実はオイラー法ではなく**修正オイラー法**で解くとうまくいくことが分かっています。
純粋な数列から少し話はずれますが、
せっかくなので紹介しておきます。

## 修正オイラー法
実は次のように離散化すると精度がよくなることが知られています。

\begin{align}
 v_{n+1}
 &=
 v_{n} - h \alpha^{2} x_{n}, \\
 x_{n+1}
 &=
 x_{n} + h v_{n+1}. \quad \text{($v_{n}$ が $v_{n+1}$ になった)}
\end{align}

これを実装して確認してみましょう。

```python
import numpy as np
import matplotlib.pyplot as plt

def harmonic_modified_euler(nt, init = (5, 0)):
    dt = t_range / (nt - 1)
    # 初期条件設定
    x = np.zeros(nt)
    v = np.zeros(nt)
    x[0] = init[0]
    v[0] = init[1]

    for i in range(1, nt):
        # ここで方程式を解いている
        v[i] = v[i-1] - dt * (omega ** 2) * x[i-1]
        x[i] = x[i-1] + dt * v[i]

    return (x, v)

# パラメータ設定
omega = 2 * np.pi
nt = 101
t_range = 2
init = (5, 0)

# 近似解
harm = harmonic_modified_euler(nt, init)
t = np.linspace(0, 2, nt)

# 厳密解
x_exact = init[0] * np.cos(- omega * t)
v_exact = - omega * init[0] * np.sin(omega * t)

# グラフ描画
plt.subplot(3, 1, 1) # 縦に 3 つグラフを並べていて、そのうちの一番上
plt.title('x-t graph')
plt.plot(np.linspace(0, 2, nt), harm[0])
plt.plot(t, x_exact)
plt.legend(['x approximation', 'x exact'])

plt.subplot(3, 1, 2)
plt.title('v-t graph')
plt.plot(np.linspace(0, 2, nt), harm[1])
plt.plot(t, v_exact)
plt.legend(['v approximation', 'v exact'])

plt.subplot(3, 1, 3)
plt.title('phase space')
plt.plot(harm[0], harm[1])

# 描画
plt.tight_layout()
```

## 結果の分析
今度の一致具合はなかなかよいです。
何より良くなったのは相空間軌道です：ちゃんと閉じてくれました。
