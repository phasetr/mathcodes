# 関数とは何か? 微分方程式のシミュレーションの観点から
今回は**関数**です。
そして数列、 漸化式、 微分、 微分方程式とつないでいく予定です。

## ベクトルの復習
前回はベクトルを紹介しました。
ベクトルは始点と終点をつなぐ矢印であり、
その矢印を辿って折れ線を作って曲線を近似するのだ、
そういう視点からの紹介でした。

## 関数は対応ルール
で、関数。
とりあえず**数と数を適当なルールで結びつける装置**のことだと思ってください。

関数のよくある説明は自動販売機です。
自動販売機にお金を入れて何かボタンを押すと押したボタンに対応した飲み物が出てきます。
この対応がまさに関数です。

### 細かい注意
うるさいことを言うと、
売り切れているときにその飲み物が出てこないので数学の関数とは正確には一致していません。
この例えは他の問題もあります。
例えば現実問題としては時々全然違う飲み物が出てくることもありえます。

ただ、今はそういう細かいことは無視して**理想的な**自動販売機を考えてください。
ちなみにこういううるさい細々としたことが気になるあなた、
実はとても数学に向いていますよ。
大学の数学だとこういう話を山ほどやります。
大学の数学もちょっと紹介する予定なので楽しみにしててください。

### 大事なこと
ここで数学として大事なのは、
押したボタンに応じて**いつでも同じ飲み物が出てくる**ことです。
同じボタンを押したのに違う飲み物が出てこられたらルールが壊れているからです。

数学で具体例を出すなら $f (x) = x^2$ を考えてみてください。
これは $x$ を入れたら必ず $x^2$ を出してくれる機械のことで、単にそれを関数と呼んでいます。
この関数を支配するルールはもちろん $x$ を $x^2$ に変換することです。

## 微分方程式も対応ルール
上で書いた「対応ルール」は本当に何でも構いません。
規則がはっきり書けないが何かルールはある、そういう謎の「ルール」でも構いません。

そしてこの講座で大事なのは**ルールとして微分方程式が使われている場合**です。
かといっていきなり微分にかっとぶのも大変なので、
もう少しクッションを入れます。

あなたは関数は実数に対して実数をあてるルールと思っているかもしれません。
実際にはもっとゆるく、何かと何かを適当に割り当てるルールくらいに思っておいてください。
例えば自然数に対して実数を割り当てる関数もあります。
そしてこれはふつう**数列**と呼ばれています。
つまり**数列も関数**なのです。

もしあなたが数列を勉強したことがあるなら、
数列は「ある規則にしたがって数を書き並べたもの」のような説明を受けなかったでしょうか?
これはまさに数列が関数だと言っています。

## 数列も対応ルール
上で数列も関数と書きました。
改めて思い出してもらうと、高校で数列は特にルールという趣が強い形で出てきます。
等差数列、等比数列や階差数列もそうですし、
**漸化式**はまさに数列を作っていくルールです。

最初にシミュレーションでやったことを思い出してもらうと、
プログラムを書いて計算していったのはまさに漸化式です。

### 注意
この講座ではふつうの中高の数学の教科書とは違って $0$ は自然数だとします。
FORTRAN のような大事な例外はあるものの、多くのプログラミング言語ではリスト・配列は 0 からはじまるので、プログラムを組むときに都合がいいのです。
少なくとも Python や C 言語のリスト・配列は $0$ からはじまります。

## ベクトルも対応ルール
あなたが大学の数学やその他の学問に興味があるなら、
次のようなことまで知っておくと楽しいかもしれません。

数列は自然数**全体**を適当なルールで実数に割り当てる関数でした。
ここで自然数全体ではなく $1$ から $n$ までを実数に割り当てる関数を考えてみましょう。
実はこれが**ベクトル**です。
$n$ 次元のベクトルというやつです。

ベクトルは他にいろいろ要件があるのであまり適当なことを言うのもよくないのですが、この視点はこの視点で大事で役に立ちます。

## 実在する高次元ベクトル
ちなみに私が大学の時、「$3$ より高い次元のベクトルなんてどこに使うの?」というのを工学系の学生が言っているのを聞いたことがあります。
実はシミュレーションでこういう高次元のベクトルをバリバリ使っています。

## 今回のポイント
最後の高次元ベクトルの話はわからなくても問題ありません。
大事なのは次の 2 つです。

- **関数は何かと何かの対応ルール**であること。
- **数列は特殊な関数**であること。

何はともあれこの 2 つをおさえてください。

あなたは数列の漸化式は数列を作るルールである、
そういう風に捉えているかもしれません。
まさにそうで、 微分方程式が関数を作るルールであるのも全く同じことです。
というわけで関数とルール設定という流れで、
次回は数列の話をします。

今回は概念の説明だけだったので少し難しかったでしょう。
次回の数列では高校の復習しながら具体的な話もします。
今回の話がわからなくてもあまり気にしないでください。

今回はここまでです。
お疲れ様でした。
