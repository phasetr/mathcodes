## SymPy によるテイラー展開
`series()` メソッドで原理的には何次まででも計算してくれる。

```python
from sympy import Symbol, cos, series
x = Symbol('x')
display(series(cos(x), x, n=2))
display(series(cos(x), x, n=4))
display(series(cos(x), x, n=6))
display(series(cos(x), x, n=8))
```

## 余弦関数 $\cos x$

```python
def plot_function_name(name, x=0, y=0):
    plt.text(x, y, name, alpha=0.3, size=25, ha="center", va="center")

figsize_x = 10
figsize_y = 10
x_min = -15
x_max = 15
y_min = -2
y_max = 2
step = 0.01
fig_row = 3
fig_col = 3

x = np.arange(x_min, x_max, step)
one = np.ones(int((x_max - x_min) / step))
fs = [one,
      one - x**2 / 2,
      one - x**2 / 2 + x**4 / 24,
      one - x**2 / 2 + x**4 / 24 - x**6 / 720,
      one - x**2 / 2 + x**4 / 24 - x**6 / 720 + x**8 / math.factorial(8),
      one - x**2 / 2 + x**4 / 24 - x**6 / 720 + x**8 / math.factorial(8) - x**10 / math.factorial(10),
      one - x**2 / 2 + x**4 / 24 - x**6 / 720 + x**8 / math.factorial(8) - x**10 / math.factorial(10) + x**12 / math.factorial(12),
      one - x**2 / 2 + x**4 / 24 - x**6 / 720 + x**8 / math.factorial(8) - x**10 / math.factorial(10) + x**12 / math.factorial(12) - x**14 / math.factorial(14),
      one - x**2 / 2 + x**4 / 24 - x**6 / 720 + x**8 / math.factorial(8) - x**10 / math.factorial(10) + x**12 / math.factorial(12) - x**14 / math.factorial(14) + x**16 / math.factorial(16),
     ]

fig = plt.figure(figsize=(figsize_x, figsize_y))

for i, f in enumerate(fs):
    subplot = fig.add_subplot(fig_row, fig_col, i+1)
    subplot.set_xlim(x_min, x_max)
    subplot.set_ylim(y_min, y_max)
    plot_function_name(f"{2*i}-th approx")
    subplot.plot(x, np.cos(x))
    subplot.plot(x, f)
```

## 指数関数 $e^x$

```python
def plot_function_name(name, x=0, y=0):
    plt.text(x, y, name, alpha=0.3, size=25, ha="center", va="center")

figsize_x = 10
figsize_y = 10
x_min = -15
x_max = 15
y_min = -10
y_max = 10
step = 0.01
fig_row = 3
fig_col = 3

x = np.arange(x_min, x_max, step)
one = np.ones(int((x_max - x_min) / step))
fs = [one,
      one + x,
      one + x + x**2 / 2,
      one + x + x**2 / 2 + x**3 / 6,
      one + x + x**2 / 2 + x**3 / 6 + x**4 / 24,
      one + x + x**2 / 2 + x**3 / 6 + x**4 / 24 + x**5 / 120,
      one + x + x**2 / 2 + x**3 / 6 + x**4 / 24 + x**5 / 120 + x**6 / 720,
      one + x + x**2 / 2 + x**3 / 6 + x**4 / 24 + x**5 / 120 + x**6 / 720 + x**7 / 5040,
      one + x + x**2 / 2 + x**3 / 6 + x**4 / 24 + x**5 / 120 + x**6 / 720 + x**7 / 5040 + x**8 / 40320,
     ]


fig = plt.figure(figsize=(figsize_x, figsize_y))

for i, f in enumerate(fs):
    subplot = fig.add_subplot(fig_row, fig_col, i+1)
    subplot.set_xlim(x_min, x_max)
    subplot.set_ylim(y_min, y_max)
    plot_function_name(f"{i}-th approx")
    subplot.plot(x, np.exp(x))
    subplot.plot(x, f)
```

## 対数関数 $\log (1+x)$

右に向けて発散しているのは $x=0$ でのテイラー展開の終息半径の兼ね合い.

```python
def plot_function_name(name, x=0, y=0):
    plt.text(x, y, name, alpha=0.3, size=25, ha="center", va="center")

figsize_x = 10
figsize_y = 10
x_min = -0.9
x_max = 2
y_min = -3
y_max = 3
step = 0.01
fig_row = 3
fig_col = 3

x = np.arange(x_min, x_max, step)
one = np.ones(int((x_max - x_min) / step))
fs = [
      x,
      x - x**2 / 2,
      x - x**2 / 2 + x**3 / 3,
      x - x**2 / 2 + x**3 / 3 - x**4 / 4,
      x - x**2 / 2 + x**3 / 3 - x**4 / 4 + x**5 / 5,
      x - x**2 / 2 + x**3 / 3 - x**4 / 4 + x**5 / 5 - x**6 / 6,
      x - x**2 / 2 + x**3 / 3 - x**4 / 4 + x**5 / 5 - x**6 / 6 + x**7 / 7,
      x - x**2 / 2 + x**3 / 3 - x**4 / 4 + x**5 / 5 - x**6 / 6 + x**7 / 7 - x**8 / 8,
      x - x**2 / 2 + x**3 / 3 - x**4 / 4 + x**5 / 5 - x**6 / 6 + x**7 / 7 - x**8 / 8 + x**9 / 9,
     ]

fig = plt.figure(figsize=(figsize_x, figsize_y))

for i, f in enumerate(fs):
    subplot = fig.add_subplot(fig_row, fig_col, i+1)
    subplot.set_xlim(x_min, x_max)
    subplot.set_ylim(y_min, y_max)
    plot_function_name(f"{i}-th approx")
    subplot.plot(x, np.log(1+x))
    subplot.plot(x, f)
```
