# 正弦関数 $\sin$

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
x = np.arange(-10, 10, 0.1)

ims = []
y = np.sin(x)
im = plt.plot(x, y, "r")
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
x = np.arange(- np.pi / 2 + 0.01, np.pi / 2 - 0.01, 0.01)

ims = []
y = np.tan(x)

# グラフの縦横比を合わせる
#plt.axes().set_aspect('equal', 'datalim')
im = plt.plot(x, y, "r")
```
