# 指数関数のグラフ

## $y=e^x$

```python
%matplotlib inline
from sympy import symbols, exp
from sympy.plotting import plot
x = symbols('x')
f = exp(x)
display(f)
plot(f, (x, -6, 6))
```

## $y=e^{-x}$

```python
%matplotlib inline
from sympy import symbols, exp
from sympy.plotting import plot
x = symbols('x')
f = exp(-x)
display(f)
plot(f, (x, -6, 6))
```

## $y = e^{-x^2}$
上 2 つと縦軸のスケールが全く違うことに注意する。

```python
%matplotlib inline
from sympy import symbols, exp
from sympy.plotting import plot
x = symbols('x')
f = exp(-x**2)
display(f)
plot(f, (x, -6, 6))
```

# 対数関数のグラフ

## $y = \log x$
`plot()` に区間の左端として 0 を渡しているのでエラーが出ている。

```python
%matplotlib inline
from sympy import symbols, log
from sympy.plotting import plot
x = symbols('x')
f = log(x)
display(f)
plot(f, (x, 0.01, 6))
```

## $y = - \log x$
`plot()` に区間の左端として 0 を渡しているのでエラーが出ている。

```python
%matplotlib inline
from sympy import symbols, log
from sympy.plotting import plot
x = symbols('x')
f = - log(x)
display(f)
plot(f, (x, 0.01, 6))
```
